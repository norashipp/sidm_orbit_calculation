```python
def spherical_coordinates(position):
    r = np.sqrt(position[0]**2+position[1]**2+position[2]**2)
    theta = np.arccos(position[2]/r)
    phi = np.arctan2(position[1],position[0])
    print 'x, y , z = ', position
    print 'r, theta, phi = ', r, theta, phi
    return r, theta, phi
```
